```python
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf  
from scipy.fft import fft,fftfreq

path =r"C:\Users\pcDK\dennryokuken\soturon\program\sound_from_harvard\ESC-50\201 - Rain\AnyConv.com__1-17367-A.wav"
data, samplerate = sf.read(path)    
print(data,samplerate)
T=1/samplerate#サンプリング周期
N=len(data)#サンプル数
print(N)
Figure=plt.figure()
figure1=Figure.add_subplot(2,1,1)
figure2=Figure.add_subplot(2,1,2)
plt.subplots_adjust(hspace=1)
#FFT↓
FFT=fft(data)
print(FFT)
f=fftfreq(N,T)[:N//2]
figure1.set_xlabel("frequency(Hz)")
figure1.set_ylabel("signal amplitude")
figure1.plot(f,2.0/N*np.abs(FFT[0:N//2]))

print(f)
FFT_abs=2.0/N*np.abs(FFT[0:N//2])
M=max(FFT_abs)
print(M)
A=np.where(FFT_abs==M)
print(A)
print(f[A])
figure2.set_xlabel("frequency(Hz)")
figure2.set_ylabel("signal amplitude")
figure2.plot(f,2.0/N*np.abs(FFT[0:N//2]))
figure2.set_xlim(0,100)
```
