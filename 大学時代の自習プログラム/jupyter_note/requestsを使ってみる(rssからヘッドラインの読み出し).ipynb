```python
import requests
rq=requests.get("https://www.yahoo.co.jp")
print(rq.text)
```

```python
import requests 
from bs4 import BeautifulSoup

xml=requests.get("https://news.yahoo.co.jp/rss/categories/it.xml")
soup=BeautifulSoup(xml.text,"html.parser")
for news in soup.findAll("item"):
    print(news.title.string)
```

```python
xmls=requests.get("http://blog.livedoor.jp/kinisoku/index.rdf")
sp=BeautifulSoup(xmls.text,"html.parser")
for n in sp.findAll("item"):
    print(n.title.string)
```
