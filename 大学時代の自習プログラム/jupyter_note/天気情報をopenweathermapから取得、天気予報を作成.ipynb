```python
import requests
import os
import datetime
from googletrans import Translator
url="https://api.openweathermap.org/data/2.5/forecast?q=tokyo&appid=e25c2cb3e57becb96db8390dfc0abad5"
data=requests.get(url).json()
#resuestsオブジェクトに対してget,jsonメソッドを実行↑
print("東京の天気予報\n")
for cityweather in data["list"]:
#citweatherは予報時間ごとの天気情報を走査
    truedatetime=str(datetime.datetime.fromtimestamp(cityweather["dt"]))
    translated=Translator().translate(cityweather["weather"][0]["main"],"ja")
    print("日時:"+truedatetime+"  天気:"+cityweather["weather"][0]["main"]
    )
```

```python
import requests
import os
import datetime

url="https://api.openweathermap.org/data/2.5/forecast?q=nagasaki&lang=ja&appid=e25c2cb3e57becb96db8390dfc0abad5"
data=requests.get(url).json()
#resuestsオブジェクトに対してget,jsonメソッドを実行↑
print("東京の天気予報\n")

for cityweather in data["list"]:
#citweatherは予報時間ごとの天気情報を走査
    truedatetime=str(datetime.datetime.fromtimestamp(cityweather["dt"]))
    print("日時:"+truedatetime+"  天気:"+cityweather["weather"][0]["description"])
```
